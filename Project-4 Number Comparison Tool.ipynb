- Number Comparison Tool (Function Based)

```python
## Compare Two Numbers
def comp(a,b):
    if a > b:
        print("A is Greater!",a)
    elif b > a:
        print("B is Greater!",b)
    else:
        print("Both A and B are Equal!",a,b)
```

```python
n1 = float(input("Enter First Number : "))
n2 = float(input("Enter Second Number : "))
result = comp(n1,n2)
```
